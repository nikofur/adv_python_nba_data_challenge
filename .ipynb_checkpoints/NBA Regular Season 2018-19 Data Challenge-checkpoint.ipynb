# What was the average age of player in the league?

```python
import pandas as pd
import numpy as np
nba_data = pd.read_csv('nbastats2018-2019.csv', sep = ',')
average_age = nba_data['Age'].mean()
print(f"The average age of a player was {average_age} in 2018-19")
```

# What player scored the most points?

```python
most_points = nba_data.sort_values('Points', kind = 'mergesort', ascending = False).reset_index()
most_points.loc[0][['Name', 'Points']]
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
most_blocks = nba_data.sort_values('Blocks', ascending = False).reset_index()
most_blocks.loc[0][['Name', 'Blocks']]

#  I dont see player positions in the data but according to google, he is a center
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
best_ws48 = nba_data[nba_data['MP'] > 20].sort_values('WS48', ascending = False).reset_index()
best_ws48.loc[0][['Name', 'MP', 'WS48']]

# Zhou Qi has the best WS48 which tells a player's win share per 48 minutes.  But, with only 1
# minute played per game, Qi was not as impactful as this stat implies.  I narrowed the query to
# only include players who average atleast 20 minutes a game as they are the ones that are going 
# to influence the outcome of games more
```

# What player had the best 3-pt percentage? 

```python
best_3fg = nba_data[(nba_data['G'] > 50) & (nba_data['MP'] > 20)].sort_values('FG3%', ascending = False).reset_index()
best_3fg[['Name', 'Team', 'G', 'MP', 'FG3%', 'FGA']]

# Out of all of the players that played in atleast 50 games and played an average of atleast 20 minutes per game,
# Domantas Sabonis had the best 3 pt. percentage in the league hitting very close to 53% of his shots
```

# Who played the most minutes during the season

```python
nba_data['Total Min Played'] = nba_data['G'] * nba_data['MP']
nba_data.sort_values('Total Min Played', ascending = False)[['Name', 'Team', 'G', 'MP', 'Total Min Played']]

#  Bradley Beal of the Washington Wizards played in all 82 games and averaged 36.9 minutes per game.  
#  This adds up to a total of 3,025.8 minutes played throughout the entire season
```

# What player given their player effiecency rating was the clutchest during the season?

```python
clutchest_players = nba_data[(nba_data['G'] > 50) & (nba_data['MP'] > 20)].sort_values('PER', ascending = False)
clutchest_players

# Out of all of the players that played in atleast 50 games and played atleast an average of 20 minutes 
# per game, Giannis Antelokounmpo of the Milwaukee Bucks had the best player efficiency rating.  Close
# behind Giannis was James Harden of the Houston Rockets and Anthony Davis of the New Orleans Pelicans
```

# What team had the youngest roster?

```python
youngest_roster = nba_data.groupby('Team', as_index = False)['Age'].mean()
youngest_roster.sort_values('Age', ascending = True)

# The Chicago Bulls had the youngest team with an average age of 24.31 years
```
